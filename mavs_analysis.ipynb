Below we evaluate our team(Dallas Mavericks) using the four factor model for both offense and defense. Offense first. Our data here comes from basketball-reference.com. Out of the 30 teams in the NBA, it is very apparent the Mavericks were average offenesively for the 2018-2019 season. We were ranked 19th in Effective Field Goal Percentage(eFG%), which adds all field goals made to 3 point field goals made times a weight of 0.5 divided by all field goals attempted. The Mavericks were also ranked 20th in Turnover Percenetage(TOV%), which is a value that estimates how many turnovers we commit per 100 plays. The Mavericks also ranked 15th in Offensive Rebound Percentage(OBP%), which a value that represents the percentage of rebounds our team gets off of our missed shots. Although, we were average in other areas, we had a pretty good Free Throw Rate(FT/FGA or FTR), which divides free throws made by field goal attempts made. We ranked 5th in the league in FTR, which means we got to the foul line a lot.

```python
offense_columns = ['eFG%', 'TOV%', 'ORB%', 'FT/FGA']
rows = ['Team', 'Lg Rank']
cell_text = [['0.519', '12.7', '22.7', '0.216'], ['19', '20', '15', '5']]

table = plt.table(cellText=cell_text,
         rowLabels=rows,
         rowColours=None,
         colLabels=offense_columns,
         loc='center')

table.auto_set_font_size(False)
table.set_fontsize(24)
table.scale(4, 4)

plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
plt.tick_params(axis='y', which='both', right=False, left=False, labelleft=False)

for pos in ['right','top','bottom','left']:
    plt.gca().spines[pos].set_visible(False)
```

```python
import pandas as pd

nba_metrics_data = pd.read_csv('basketball_adv_metrics_18_19.csv')
nba_metrics_data.dropna(inplace=True)
nba_metrics_data.set_index('Tm', inplace=True)
nba_metrics_data.rename(columns={'eFG%.1': 'DeFG%', 'TOV%.1':'DTOV%', 'ORB%.1':'DRB%', 'FT/FGA.1': 
                                 'Def FT/FGA'},  inplace=True)
nba_metrics_data['DRB%'] = 100 - nba_metrics_data['DRB%']
nba_metrics_data
```

```python
from sklearn.model_selection import train_test_split

X = nba_metrics_data.drop(['W', 'L'], axis=1)
y = nba_metrics_data.loc[:, 'W']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha=0.1)
ridge_reg.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(ridge_reg, X_train, y_train, cv=10, scoring='neg_mean_squared_error')
```

```python
y_pred = ridge_reg.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, y_pred)
```

Below we see the most important we see that the moast important feature in predicting wins is a team's offensive Effective Field Goal Percentage(eFG%). It has the largest coefficient.

```python
ridge_reg.coef_
```

Again, we see visually that our team's eFG% is the best predictor of team wins.

```python
plt.plot(X.columns, ridge_reg.coef_, 'b.')
plt.axis([-1, 8, -30, 30])
plt.show()
```

Now, we predict how many games we should've expected to win given our model. According to our model the Mavericks were expected to win 41 games. However, we only won 33 games last season. So, we didn't meet expectations. My suspicion is we underachieved due to trying to tank to get a better draft pick. We could probably find that out by adding in the minutes played by our starters to our model. We also traded away Harrison Barnes and Dennis Smith before the trade deadline. We could examine if our eFG% decreased after te aformentioned players to test if this hypothesis is true.
